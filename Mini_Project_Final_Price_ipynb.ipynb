```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Load Both Dataset
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")
```

```python
x = train_df[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]

y = train_df['price_range']
```

```python
# This line divides the dataset (x and y) into four parts — training and testing sets — where 80% of the data is used to train the model (x_train, y_train) and 20% is used to evaluate its accuracy (x_test, y_test).

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 ,random_state = 12,)
```

```python
# Call required models
model_knn = KNeighborsClassifier()
model_NB = GaussianNB()
model_LR = LogisticRegression()
model_svm = SVC()
```

```python
# Each line trains a different machine learning model
model_knn.fit(x_train,y_train)
model_NB.fit(x_train,y_train)
model_LR.fit(x_train,y_train)
model_svm.fit(x_train,y_train)
```

```python
# Each line tells its model: Now that you’re trained, predict the outputs (y values) for the unseen test inputs (x_test).

y_pred_knn = model_knn.predict(x_test)
y_pred_NB = model_NB.predict(x_test)
y_pred_LR = model_LR.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_knn))
```

```python
print(classification_report(y_test,y_pred_NB))
```

```python
print(classification_report(y_test,y_pred_LR))
```

```python
print(classification_report(y_test,y_pred_svm))
```
